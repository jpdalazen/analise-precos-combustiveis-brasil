```python
import pandas as pd

df = pd.read_csv('dadosgasolina.tsv', sep='\t')
df.head()
```


A seguir, analisamos a distribuição do preço médio de revenda da gasolina.

Essa visualização permite identificar:
- Concentração dos preços
- Possíveis outliers
- Faixa mais comum de valores praticados

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
sns.histplot(df['PREÇO MÉDIO REVENDA'], bins=10)
plt.title('Distribuiçao do Preço Médio de Revenda')
plt.xlabel('Preço Médio  (R$)')
plt.ylabel('Frequencia')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PREÇO MÍNIMO REVENDA', y='PREÇO MÁXIMO REVENDA', data=df)
plt.title('Preço Mínimo vs Preço Máximo de Revenda')
plt.xlabel('Preço Mínimo (R$)')
plt.ylabel('Preço Máximo (R$)')
plt.show()
```

```python
df.dtypes
```

```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'])
df['DATA FINAL'] = pd.to_datetime(df['DATA FINAL'])
```

Nesta análise, calculamos o preço médio de revenda da gasolina por estado.

Em seguida, ordenamos os resultados para identificar os estados com os maiores preços médios.

```python
preco_por_estadoc = (
    df.groupby('ESTADO')['PREÇO MÉDIO REVENDA']
    .mean()
    .sort_values(ascending=False)
)

preco_por_estadoc.head(27)
```

```python
plt.figure(figsize=(10,5))
preco_por_estadoc.head(10).plot(kind='bar', color='red')
plt.title('Top 10 Estados com Maior Preço Médio de Revenda de Gasolina')
plt.ylabel('Preço Médio (R$)')
plt.xlabel('Estado')
plt.xticks(rotation=45)

plt.show()
```

```python
preco_por_estadob = (
    df.groupby('ESTADO')['PREÇO MÉDIO REVENDA']
    .mean()
    .sort_values(ascending=True)
)

preco_por_estadob.head(10)
```

```python
plt.figure(figsize=(10,5))
preco_por_estadob.head(10).plot(kind='bar', color='green')
plt.title('Top 10 Estados com Menor Preço Médio de Revenda de Gasolina')
plt.ylabel('Preço Médio (R$)')
plt.xlabel('Estado')
plt.xticks(rotation=45)
plt.show()
```

```python
volatilidade_estado = (
    df.groupby('ESTADO')['PREÇO MÉDIO REVENDA']
    .std()
    .sort_values(ascending=False)
)

volatilidade_estado.head(27)
```

```python
preco_por_regiao = (
    df.groupby('REGIÃO')['PREÇO MÉDIO REVENDA']
    .mean()
    .sort_values(ascending=False)
)
preco_por_regiao
```

Aqui avaliamos o preço médio da gasolina agrupado por região do país.

Essa análise ajuda a entender diferenças estruturais de preços relacionadas a logística, impostos e oferta regional.

```python
plt.figure(figsize=(10,5))
preco_por_regiao.head(5).plot(kind='bar', color='purple')
plt.title('Preço Medio por Regiao')
plt.ylabel('Preço Médio (R$)')
plt.xlabel('Regiao')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x=preco_por_regiao.index,
    y=preco_por_regiao.values,
    palette='viridis'
)
plt.title('Preço Médio da Gasolina por Regiao')
plt.ylabel('Preço Médio (R$)')
plt.xlabel('Regiao')
plt.show()
```

```python
df['DATA INICIAL'] = pd.to_datetime(df['DATA INICIAL'])
```

```python
preco_tempo = (
    df.groupby('DATA INICIAL')['PREÇO MÉDIO REVENDA']
    .mean()
)

preco_tempo.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(preco_tempo.index, preco_tempo.values)
plt.title('Evoluçao do Preço Médio da Gasolina no Brasil')
plt.xlabel('Ano')
plt.ylabel('Preço Médio (R$)')
plt.show()
```

Nesta etapa analisamos a evolução do preço médio da gasolina ao longo do tempo.

O objetivo é identificar tendências, períodos de alta ou queda e possíveis impactos econômicos ou políticos nos preços.

```python
df['PRODUTO'].value_counts()
```

```python
ordem = (
    df.groupby('PRODUTO')['PREÇO MÉDIO REVENDA']
    .median()
    .sort_values()
    .index
)

plt.figure(figsize=(10,5))
sns.boxplot(
    data=df,
    x='PRODUTO',
    y='PREÇO MÉDIO REVENDA',
    order=ordem
)
plt.title('Distribuição de preços por tipo de combustível')
plt.xticks(rotation=30)
plt.show()
```

```python
df['DIF_REVENDA_DISTRIB'] = (
    df['PREÇO MÉDIO REVENDA'] - df['PREÇO MÉDIO DISTRIBUIÇÃO']
)
df['DIF_REVENDA_DISTRIB'].describe()
```

```python
df[['PREÇO MÉDIO REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO']].dtypes
```

```python
df['PREÇO MÉDIO DISTRIBUIÇÃO'] = (
    df['PREÇO MÉDIO DISTRIBUIÇÃO']
    .astype(str)
    .str.replace(',','.',regex=False)
)
df['PREÇO MÉDIO DISTRIBUIÇÃO'] = pd.to_numeric(
    df['PREÇO MÉDIO DISTRIBUIÇÃO'],
    errors='coerce'
)
```

Nesta análise calculamos a diferença entre o preço médio de revenda e o preço médio de distribuição.

Essa métrica ajuda a compreender a margem existente entre distribuição e venda final ao consumidor.

```python
df['DIF_REVENDA_DISTRIB'] = (
    df['PREÇO MÉDIO REVENDA'] - df['PREÇO MÉDIO DISTRIBUIÇÃO']
)

df['DIF_REVENDA_DISTRIB'].describe()
```
